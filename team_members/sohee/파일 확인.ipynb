```python
import os
import pandas as pd

# 1. CSV의 파일명 리스트
df = pd.read_csv('data/raw/data_list.csv')
csv_filenames = set(df['파일명'].tolist())
print(f"CSV에 있는 파일명 개수: {len(csv_filenames)}")

# 2. 실제 업로드된 파일 리스트
files_path = 'data/raw/files'
uploaded_files = os.listdir(files_path)
print(f"업로드된 파일 개수: {len(uploaded_files)}")

# 3. "의 사본" 파일 개수 확인
duplicate_files = [f for f in uploaded_files if '의 사본' in f]
print(f"\n'의 사본' 파일 개수: {len(duplicate_files)}")

# 4. 원본 파일만 있는지 확인
original_files = [f for f in uploaded_files if '의 사본' not in f]
print(f"원본 파일 개수: {len(original_files)}")

# 5. CSV 파일명과 매칭되는 파일 확인
matched_files = [f for f in uploaded_files if f in csv_filenames]
print(f"\nCSV와 매칭되는 파일 개수: {len(matched_files)}")
```

```python
import os

files_path = 'data/raw/files'
uploaded_files = os.listdir(files_path)

# "의 사본" 파일 찾기
duplicate_files = [f for f in uploaded_files if '의 사본' in f]

print(f"삭제할 파일 개수: {len(duplicate_files)}")
print("\n처음 5개:")
for f in duplicate_files[:5]:
    print(f"  - {f}")

# 삭제 실행
print("\n삭제 중...")
for f in duplicate_files:
    file_path = os.path.join(files_path, f)
    os.remove(file_path)
    
print(f"✅ {len(duplicate_files)}개 파일 삭제 완료!")

# 확인
remaining_files = os.listdir(files_path)
print(f"남은 파일 개수: {len(remaining_files)}")
```
